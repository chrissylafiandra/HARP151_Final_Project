The first two arguments get assigned to the x and y attributes of the rect, the coordinates of the topleft corner. So if you want to create a rect that is positioned at the coordinates (200, 300), you can write: rect = pygame.Rect(200, 300, 40, 50). The third and fourth argument are indeed the dimensions of the resulting rect instance.


n Pygame, the .rect attribute is a crucial part of working with rectangular areas. Letâ€™s dive into what it represents:

Rect Objects:
Pygame uses Rect objects to store and manipulate rectangular areas.
A Rect can be created using a combination of the following values:
left: The x-coordinate of the top-left corner.
top: The y-coordinate of the top-left corner.
width: The width of the rectangle.
height: The height of the rectangle.


Common Uses:
Rect objects are used for:
Storing the blit position of sprites or images.
Collision detection (e.g., with colliderect or collidepoint methods).
Remember, the first two arguments of a Rect represent the x and y coordinates of the top-left corner, allowing you to position the rectangle precisely on the screen1.

```python
#VARIABLES


TILESIZE = 13
PC_INIT_X = 600
PC_INIT_Y = 600
PLAYER_SPEED = 5
NPC_INIT_X = 500
NPC_INIT_Y = 700
NPC_SPEED = 4
PLAYER_LAYER = 4
NPC_LAYER = 2
WALL_LAYER = 3


NPC2_SPEED = 4
NPC2_LAYER = 3
NPC2_INIT_X = 300
NPC2_INIT_Y = 500


NPC3_SPEED = 4
NPC3_LAYER = 3
NPC3_INIT_X = 700
NPC3_INIT_Y = 500


NPC4_SPEED = 4
NPC4_LAYER = 3
NPC4_INIT_X = 600
NPC4_INIT_Y = 500


NPC5_SPEED = 4
NPC5_LAYER = 3
NPC5_INIT_X = 400
NPC5_INIT_Y = 300


BOX_INIT_X = 800
BOX_INIT_Y = 600
BOX_LAYER = 1

ROBOT_INIT_X = 50
ROBOT_INIT_Y = 550

OBJECTIVE_INIT_X = 780
OBJECTIVE_INIT_Y = 0
OBJECTIVE_LAYER = 1


SCREEN_WIDTH = 1067
SCREEN_HEIGHT = 779


#playing comes into use for the menu screen
playing = False


#WALLS ARE PARAMETERIZED BY X,Y, width height
#coordinates on pygame starts in top left corner as (0,0), down axis is y in positive and side axis is x and positive


#joke = requests.get("https://official-joke-api.appspot.com/random_joke")
#joke_json = joke.json()
#setup = joke_json["setup"]
#punchline = joke_json["punchline"]


#FIGURE OUT WHERE TO PUT CODE FOR INTRO SCREEN WHEN CELESTE FINISHES
#CLASS FOR ENTIRE GAME, EVERY OTHER CLASS/FUNCTION WORKING WITH THIS!!!!


def resize(image, max_width, max_height):
    '''made this function to resize the images while preserving the aspect ratio more easily,
    since pygame doesn't have a built in function for this
    '''
    original_width, original_height = image.get_size()
    #use tuple unpacking to set the width & height equal to the dimensions of the image


    #takes the minimum between the two quotients of the max width you'd want & the max height to be the scale
    scale = min(max_width / original_width, max_height / original_height)


    # Calculate new dimensions
    new_width = int(original_width * scale)
    new_height = int(original_height * scale)


    # Resize the image
    return pygame.transform.scale(image, (new_width, new_height))
   
intro_screen = pygame.image.load("C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\intro_screen.png")
intro_screen = pygame.transform.scale(intro_screen,(SCREEN_WIDTH,SCREEN_HEIGHT))


def get_font(size):


    font_text = "C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\joystix.otf"
    return pygame.font.Font(font_text, size)


def menu():
    global playing
    pygame.display.set_caption("MENU")
               
    while playing == False:


        #Creating all of the text
        menu_font = get_font(80).render("CRASH LANDING", True, "#68FFE9")
        menu_rect = menu_font.get_rect(center=(SCREEN_WIDTH//2, SCREEN_HEIGHT//2 - 100))
        drop_shadow = get_font(80).render("CRASH LANDING", True, "#025660")
        drop_shadow_rect = drop_shadow.get_rect(center=(SCREEN_WIDTH//2 - 5, SCREEN_HEIGHT//2 - 105))


        press_space = get_font(30).render("> PRESS SPACE TO START", True, "#68FFE9")
        space_rect = press_space.get_rect(center=(SCREEN_WIDTH//2, SCREEN_HEIGHT//2 + 200))


        #placing all of the text in the correct order
        Game.screen.blit(intro_screen,(0,0))
        intro_screen.blit(drop_shadow, drop_shadow_rect)
        intro_screen.blit(menu_font, menu_rect)
        intro_screen.blit(press_space, space_rect)


        #checking to see if the player is hitting X or if they're hitting spacebar
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
            if event.type == pygame.KEYDOWN:
                if event.key == pygame.K_SPACE:
                    playing = True
                    #playing was False before
                   
        pygame.display.flip()


class Game:
    screen = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
    def __init__(self):
        pygame.init()
        pygame.display.set_caption('rpg test')
        self.clock = pygame.time.Clock()
        self.all_sprites = pygame.sprite.LayeredUpdates()
        self.wallup = Wall(0, 0, 3, 979)
        self.wallleft = Wall(0, 0, 1340, 50)
        self.wallright = Wall(1067, 0, 50, 900)
        self.walldown = Wall(0, 779, 2000, 500)
        self.wallupleft = Wall(0, 0, 300, 300)
        self.wallupright = Wall(760, 30, 800, 290) #(MOVES RIGHTWARDS, MOVES DOWNWARDS, WIDTH, HEIGHT)
        self.player = Player(self)
        self.NPC = NPC(self)
        self.NPC2 = NPC2(self)
        self.NPC3 = NPC3(self)
        self.NPC4 = NPC4(self)
        self.NPC5 = NPC5(self)
        self.box = Box(self)
        self.objective = Objective(self)
        self.robot = Robot(self)
        
        #self.objective_text = "The Objective is to..."
       
        #LOADS BACKGROUND IMAGES:
        self.background = pygame.image.load("C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\spaceship_interior.png")
        self.background = pygame.transform.scale(self.background, (SCREEN_WIDTH, SCREEN_HEIGHT))
        self.text_box = pygame.image.load("C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\text_box.png")
       
        self.clicked_text = None #need this so text doesnt automatically fill screen when gui loads
        self.clicked_text2 = None
        self.clicked_text_robot = None
        
        self.clicked_image = pygame.image.load("C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\fusion_core.png")
        self.clicked_image = resize(self.clicked_image, 50, 50)
        self.clicked_image_rect = self.clicked_image.get_rect()
       
        self.end_pic = pygame.image.load("C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\finish_screen.png")
        self.end_pic_rect = self.end_pic.get_rect()
       
        self.clicked_image = None #definitely need this so image doesnt automatically fill screen when gui loads
        self.end_pic = None
                        #FIGURE OUT HOW TO EXIT AFTER THIS AND PUT A CONGRATS IMAGE...
       
   
    #CREATES EVENT FOR MOUSE CLICKS ON OUR NPCS
    def event(self):
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                pygame.quit()
                sys.exit()
            elif event.type == pygame.MOUSEBUTTONDOWN:
       
                if event.button == 1:  # clicks left side of mouse button
                    mouse_position = pygame.mouse.get_pos()
                    if self.NPC2.rect.collidepoint(mouse_position):
                        self.clicked_text = "Oh yeah I have the fusion core, here you go!"
                        self.clicked_text2 = None
                        self.clicked_text_robot = None
                        self.clicked_image = pygame.image.load("C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\fusion_core.png")
                        self.clicked_image_rect = self.clicked_image.get_rect()
                       
                        self.end_pic = pygame.image.load("C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\finish_screen.png")
                        self.end_pic_rect = self.end_pic.get_rect()
                        
                        #QUITS GAME AFTER 20 SECONDS when npc clicked, is in terms of milliseconds for some reason....
                       
                        pygame.time.set_timer(pygame.QUIT, 20000)
                        if event.type == pygame.QUIT:
                            pygame.QUIT
                           
                        #FIGURE OUT HOW TO EXIT AFTER THIS AND PUT A CONGRATS IMAGE...
                   
                    elif self.box.rect.collidepoint(mouse_position):
                        self.joke = requests.get("https://official-joke-api.appspot.com/random_joke")
                        self.joke_json = self.joke.json()
                        self.setup = self.joke_json["setup"]
                        self.punchline = self.joke_json["punchline"]
                        self.clicked_text_robot = None
                        
                        self.clicked_text2 = f"{self.setup}, {self.punchline}"
                        self.clicked_image = None
                        self.clicked_text = None
                       
                    else:
               
                        if self.NPC.rect.collidepoint(mouse_position):
                            self.clicked_text2 = None
                            self.clicked_text_robot = None
                            self.clicked_text = f"Why do you think I know where the fusion core is? Leave me alone..."
                            self.clicked_image = None
                           
                        elif self.NPC3.rect.collidepoint(mouse_position):
                            self.clicked_text2 = None
                            self.clicked_text_robot = None
                            self.clicked_text = "I don't have the fusion core! Go bother someone else..."
                            self.clicked_image = None
                            
                        elif self.robot.rect.collidepoint(mouse_position):
                            self.clicked_text2 = None
                            self.clicked_text = None
                            self.clicked_text_robot = None
                            
                            self.pun = requests.get("https://punapi.rest/api/pun")
                            self.pun_json = self.pun.json()
                            self.pun_text = f"{self.pun_json['pun']}"
                            self.clicked_text_robot = self.pun_text
                            
                            self.clicked_image = None
                           
                        elif self.NPC4.rect.collidepoint(mouse_position):
                            self.clicked_text2 = None
                            self.clicked_text_robot = None
                            self.clicked_text = "What the hell is a fusion core? I'm the ship therapist!"
                            self.clicked_image = None
                           
                        elif self.NPC5.rect.collidepoint(mouse_position):
                            self.clicked_text2 = None
                            self.clicked_text_robot = None
                            self.clicked_text = "I think I saw the fusion core floating around here somewhere... What's happening again?"
                            self.clicked_image = None
                       
                           
    #DISPLAYS TEXT FROM MOUSE CLICKS          
    def display_text(self, text, color, position, bg_color):
        font_text = "C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\joystix.otf"
        font = pygame.font.Font(font_text, 13)
        text_surface = font.render(text, True, color)
        text_rect = text_surface.get_rect()
        text_rect.topleft = position
       
        if bg_color: #creates rectangle to put text into
            pygame.draw.rect(self.screen, bg_color, (text_rect.left - 30, text_rect.top - 5, text_rect.width + 50, text_rect.height + 20))
       
        self.screen.blit(text_surface, text_rect) #BLITS TEXT ONTO SCREEN
       


 #updates sprites
    def update(self):
        self.all_sprites.update()
     
 #updates background, we need this so player and sprites dont create a trail behind them
    def show_change(self):
        self.screen.blit(self.background, (0, 0)) #blit is like .pack in tkinter, we're blitting background image
        self.all_sprites.draw(self.screen) #draws sprite onto screen
        if self.clicked_text:
            self.display_text(self.clicked_text, "white", (50, 50), (83, 125, 154))  # Display text with black background
       
        if self.clicked_text2:
            self.display_text(self.clicked_text2, "white", (10,100), (83, 125, 154))
            
        if self.clicked_text_robot:
            self.display_text(self.clicked_text_robot, "white", (10,100), (83, 125, 154))
           
        if self.clicked_image:
            self.screen.blit(self.clicked_image, (150,150))
           
        if self.end_pic:
            self.screen.blit(self.end_pic, (100,240))
           
        #if self.objective_text:
            #self.screen.blit(self.objective_text, (700,50))
       


        pygame.display.update()


    def run(self): #RUNS GAME
        #if you haven't yet hit spacebar in the menu screen, playing is initialized to false, and the menu function is run instead
        if playing == False:
            menu()


        #as soon as you hit the spacebar, playing is set to true, and the game runss
        if playing == True:
            while True:
                self.event()
                self.update()
                self.show_change()
                self.clock.tick(60)


#movement, wall collision and image for player
#Sprite is a base class from pygame.sprite
class Player(pygame.sprite.Sprite):
    def __init__(self, game):
        super().__init__()
        self.game = game
        self.layer = PLAYER_LAYER
        self.image = pygame.image.load("C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\main_character.png").convert_alpha()
        self.image = resize(self.image, 40, 40)
        self.image.set_colorkey((255,255,255))
        self.rect = self.image.get_rect(x=PC_INIT_X, y=PC_INIT_Y)
        self.game.all_sprites.add(self)
        self.speed = PLAYER_SPEED


    def move_player(self):
        keys = pygame.key.get_pressed()
        self.movex = (keys[pygame.K_RIGHT] - keys[pygame.K_LEFT]) * self.speed
        self.movey = (keys[pygame.K_DOWN] - keys[pygame.K_UP]) * self.speed


        self.rect.x += self.movex
        self.rect.y += self.movey


        if self.rect.colliderect(self.game.wallup.rect) or self.rect.colliderect(self.game.wallleft.rect) or self.rect.colliderect(self.game.wallupleft.rect) or self.rect.colliderect(self.game.wallupright.rect) or self.rect.colliderect(self.game.wallright.rect) or self.rect.colliderect(self.game.walldown.rect):
            self.rect.x -= self.movex
            self.rect.y -= self.movey


    def update(self):
        self.move_player()


#NOTE: WE USE .convert_alpha() for images with transparent backgrounds to hold their transparency on GUI
#movements and image for NPC1
class NPC(pygame.sprite.Sprite):
    def __init__(self, game):
        super().__init__()
        self.game = game
        self.layer = NPC_LAYER
        self.image = pygame.image.load("C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\small_spaceman_clear.png").convert_alpha()
        self.image = resize(self.image, 40,40)
        self.rect = self.image.get_rect(x=NPC_INIT_X, y=NPC_INIT_Y)
        self.game.all_sprites.add(self)
        self.speed = NPC_SPEED


    def move_NPC(self):
        movement_options = [0, TILESIZE, -TILESIZE]
        movement_weight = [99, 1, 1]


        #parameterizes movement in x and y, then below we use the premethods of .rect.x and .rect.y of npc and .movex
        #.movey to set npc movement
        self.movex = random.choices(movement_options, movement_weight)[0] * self.speed
        self.movey = random.choices(movement_options, movement_weight)[0] * self.speed


        self.rect.x += self.movex
        self.rect.y += self.movey
       
        #should block npc from leaving gui but it catches them instead, need to fix npc movements to solve issue
        if self.rect.colliderect(self.game.wallup.rect) or self.rect.colliderect(self.game.wallleft.rect) or self.rect.colliderect(self.game.wallupleft.rect) or self.rect.colliderect(self.game.wallupright.rect) or self.rect.colliderect(self.game.wallright.rect) or self.rect.colliderect(self.game.walldown.rect):
            #self.rect.x -= self.movex
            #self.rect.y -= self.movey
            self.rect.x = NPC_INIT_X
            self.rect.y = NPC_INIT_Y


    def update(self):
        self.move_NPC()
       
 #movements and image for NPC2
class NPC2(pygame.sprite.Sprite):
    def __init__(self, game):
        super().__init__()
        self.game = game
        self.layer = NPC2_LAYER
        self.image = pygame.image.load("C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\small_spaceman_clear.png").convert_alpha()
        self.image = resize(self.image, 40,40)
        self.rect = self.image.get_rect(x=NPC2_INIT_X, y=NPC2_INIT_Y)
        self.game.all_sprites.add(self)
        self.speed = NPC2_SPEED


    def move_NPC2(self):
        movement_options = [0, TILESIZE, -TILESIZE]
        movement_weight = [50, 0.5, 0.5]


        self.movex = random.choices(movement_options, movement_weight)[0] * self.speed #initializes movement in x
        self.movey = random.choices(movement_options, movement_weight)[0] * self.speed #initializes movement in y
       
        #should block npc from leaving map, sort of catches them instead, need to fix npc movement to work properly
        if self.rect.colliderect(self.game.wallup.rect) or self.rect.colliderect(self.game.wallleft.rect) or self.rect.colliderect(self.game.wallupleft.rect) or self.rect.colliderect(self.game.wallupright.rect) or self.rect.colliderect(self.game.wallright.rect) or self.rect.colliderect(self.game.walldown.rect):
            #self.rect.x -= self.movex
            #self.rect.y -= self.movey
            self.rect.x = NPC2_INIT_X
            self.rect.y = NPC2_INIT_Y


        self.rect.x += self.movex #moves npc in x
        self.rect.y += self.movey #moves npc in y
   
    def update(self):
        self.move_NPC2()
       

    
class NPC3(pygame.sprite.Sprite):
    def __init__(self, game):
        super().__init__()
        self.game = game
        self.layer = NPC3_LAYER
        self.image = pygame.image.load("C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\small_spaceman_clear.png").convert_alpha()
        self.image = resize(self.image, 40,40)
        self.rect = self.image.get_rect(x=NPC3_INIT_X, y=NPC3_INIT_Y)
        self.game.all_sprites.add(self)
        self.speed = NPC3_SPEED


    def move_NPC3(self):
        movement_options = [0, TILESIZE, -TILESIZE]
        movement_weight = [50, 0.5, 0.5]


        self.movex = random.choices(movement_options, movement_weight)[0] * self.speed #initializes movement in x
        self.movey = random.choices(movement_options, movement_weight)[0] * self.speed #initializes movement in y
       
        #should block npc from leaving map, sort of catches them instead, need to fix npc movement to work properly
        if self.rect.colliderect(self.game.wallup.rect) or self.rect.colliderect(self.game.wallleft.rect) or self.rect.colliderect(self.game.wallupleft.rect) or self.rect.colliderect(self.game.wallupright.rect) or self.rect.colliderect(self.game.wallright.rect) or self.rect.colliderect(self.game.walldown.rect):
            #self.rect.x -= self.movex
            #self.rect.y -= self.movey
            self.rect.x = NPC3_INIT_X
            self.rect.y = NPC3_INIT_Y


        self.rect.x += self.movex #moves npc in x
        self.rect.y += self.movey #moves npc in y
       
    def update(self):
        self.move_NPC3()
       
    
class NPC4(pygame.sprite.Sprite):
    def __init__(self, game):
        super().__init__()
        self.game = game
        self.layer = NPC4_LAYER
        self.image = pygame.image.load("C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\small_spaceman_clear.png").convert_alpha()
        self.image = resize(self.image, 40,40)
        self.rect = self.image.get_rect(x=NPC4_INIT_X, y=NPC4_INIT_Y)
        self.game.all_sprites.add(self)
        self.speed = NPC4_SPEED


    def move_NPC4(self):
        movement_options = [0, TILESIZE, -TILESIZE]
        movement_weight = [50, 0.5, 0.5]


        self.movex = random.choices(movement_options, movement_weight)[0] * self.speed #initializes movement in x
        self.movey = random.choices(movement_options, movement_weight)[0] * self.speed #initializes movement in y
       
        #should block npc from leaving map, sort of catches them instead, need to fix npc movement to work properly
        if self.rect.colliderect(self.game.wallup.rect) or self.rect.colliderect(self.game.wallleft.rect) or self.rect.colliderect(self.game.wallupleft.rect) or self.rect.colliderect(self.game.wallupright.rect) or self.rect.colliderect(self.game.wallright.rect) or self.rect.colliderect(self.game.walldown.rect):
            #self.rect.x -= self.movex
            #self.rect.y -= self.movey
            self.rect.x = NPC4_INIT_X
            self.rect.y = NPC4_INIT_Y


        self.rect.x += self.movex #moves npc in x
        self.rect.y += self.movey #moves npc in y
   
    def update(self):
        self.move_NPC4()
       

    
class NPC5(pygame.sprite.Sprite):
    def __init__(self, game):
        super().__init__()
        self.game = game
        self.layer = NPC5_LAYER
        self.image = pygame.image.load("C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\small_spaceman_clear.png").convert_alpha()
        self.image = resize(self.image, 40,40)
        self.rect = self.image.get_rect(x=NPC5_INIT_X, y=NPC5_INIT_Y)
        self.game.all_sprites.add(self)
        self.speed = NPC5_SPEED


    def move_NPC5(self):
        movement_options = [0, TILESIZE, -TILESIZE]
        movement_weight = [50, 0.5, 0.5]


        self.movex = random.choices(movement_options, movement_weight)[0] * self.speed #initializes movement in x
        self.movey = random.choices(movement_options, movement_weight)[0] * self.speed #initializes movement in y
       
        self.rect.x += self.movex #moves npc in x
        self.rect.y += self.movey #moves npc in y
       
        #should block npc from leaving map, sort of catches them instead, need to fix npc movement to work properly
        if self.rect.colliderect(self.game.wallup.rect) or self.rect.colliderect(self.game.wallleft.rect) or self.rect.colliderect(self.game.wallupleft.rect) or self.rect.colliderect(self.game.wallupright.rect) or self.rect.colliderect(self.game.wallright.rect) or self.rect.colliderect(self.game.walldown.rect):
            #self.rect.x -= self.movex
            #self.rect.y -= self.movey
           
            self.rect.x = NPC5_INIT_X
            self.rect.y = NPC5_INIT_Y
       
    def update(self):
        self.move_NPC5()


class Box(pygame.sprite.Sprite):
    def __init__(self, game):
        super().__init__()
        self.game = game
        self.layer = BOX_LAYER
        self.image = pygame.image.load("C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\humor_box.png").convert_alpha()
        self.image = resize(self.image, 80,100)
        self.rect = self.image.get_rect(x=BOX_INIT_X, y=BOX_INIT_Y)
        self.game.all_sprites.add(self)

class Robot(pygame.sprite.Sprite):
    def __init__(self, game):
        super().__init__()
        self.game = game
        self.layer = BOX_LAYER
        self.image = pygame.image.load("C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\robot.png").convert_alpha()
        self.image = resize(self.image, 60,80)
        self.rect = self.image.get_rect(x=ROBOT_INIT_X, y=ROBOT_INIT_Y)
        self.game.all_sprites.add(self)

class Objective(pygame.sprite.Sprite):
    def __init__(self, game):
        super().__init__()
        self.game = game
        self.layer = OBJECTIVE_LAYER
        self.image = pygame.image.load("C:\\Users\\chris\\OneDrive\\Documents\\HARP_151_PROJ_PYGAME\\objective_text2.jpeg").convert_alpha()
        self.rect = self.image.get_rect(x=OBJECTIVE_INIT_X, y=OBJECTIVE_INIT_Y)
        self.game.all_sprites.add(self)
   
  #CREATES WALL AKA BOUNDS THAT DONT LET PEOPLE LEAVE REGION
class Wall(pygame.sprite.Sprite):
    def __init__(self, x, y, width, height):
        super().__init__()
        self.image = pygame.Surface((width, height))
        self.image.fill((255, 255, 255))
        self.rect = self.image.get_rect(topleft=(x, y))




Game().run()


#WORK CITED:
#General syntax: https://www.pygame.org/docs/ref/font.html/
#font: https://stackoverflow.com/questions/38001898/what-fonts-can-i-use-with-pygame-font-font
#loading images into sprite/syntax help: https://stackoverflow.com/questions/47082209/pygame-loading-images-in-sprites
#understanding .rect method: https://stackoverflow.com/questions/47829248/pygame-what-is-the-rect-arguments-exactly-and-what-do-they-represent
#how to detect mouse click: https://stackoverflow.com/questions/10990137/pygame-mouse-clicking-detection
#PYGAME DOCUMENTATION: https://www.pygame.org/docs/
```
